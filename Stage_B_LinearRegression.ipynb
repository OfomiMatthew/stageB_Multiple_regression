```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('energydata_complete.csv')
data
```

```python
#dropping the date and lights columns
modified_data = data.drop(['date','lights'], axis=1)
modified_data
```

```python
#checking wholeness of data
modified_data.info()
#no missing values in dataset
```

```python
correlation = modified_data.corr()
print(correlation)
```

```python
corr_matrix =correlation['Appliances'].sort_values(ascending = False)
corr_matrix
```

```python
#Performing scaling of the dataset
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
```

```python
normalized_data = pd.DataFrame(scalar.fit_transform(modified_data), columns=modified_data.columns)
X= normalized_data.iloc[:,1:].values
Y = normalized_data.iloc[:,0].values
```

```python
DataFrame(X)
```

```python
Series(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
DataFrame(X_train)
```

```python
DataFrame(X_test)
```

```python
Series(Y_train)
```

```python
Series(Y_test)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
```

```python
y_pred = lin_reg.predict(X_test)
Series(y_pred)
```

```python
#Evaluating on the test set
from sklearn.metrics import mean_squared_error
mse =round( mean_squared_error(y_pred,Y_test),3)
mse
```

```python
#residual sum of squared error
rss =round( np.sum(np.square(Y_test-y_pred)) , 3)
rss
```

```python
#coefficient of determination
from sklearn.metrics import r2_score
r2 = round(r2_score(Y_test,y_pred),3)
r2
```

```python
#root mean squared error
rmse = round(np.sqrt(mse),3)
rmse
```

```python
# Applying regularization(penalization) models below:
```

```python
#Ridge regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X_train,Y_train)
```

```python
y_pred = ridge_reg.predict(X_test)
Series(y_pred)
```

```python
#Lasso regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train,Y_train)
```

```python
y_pred = lasso_reg.predict(X_test)
Series(y_pred)
```

```python
#Elastic Net
from sklearn.linear_model import ElasticNet
net_reg = ElasticNet(alpha=0.1,l1_ratio=0.5)
net_reg.fit(X_train,Y_train)
```

```python
y_pred = net_reg.predict(X_test)
Series(y_pred)
```

```python
#checking for coefficients of predictorsfor each models
#lin_reg_weights
lin_reg_weights = lin_reg.coef_
Series(lin_reg_weights)
```

```python
#ridge_reg_weight
ridge_reg_weights = ridge_reg.coef_
Series(ridge_reg_weights)
```

```python
#lasso_reg_weights
lasso_reg_weights = lasso_reg.coef_
Series(lasso_reg_weights)
```

```python
#net_reg_weights
net_reg_weights = net_reg.coef_
Series(net_reg_weights)
```

```python
data_weights = DataFrame({'lin_reg_weights':lin_reg_weights, 'ridge_reg_weights':ridge_reg_weights,
                          'lasso_reg_weights':lasso_reg_weights,'net_reg_weights':net_reg_weights}, 
                           index=['T1','RH_1','T2','RH_2','T3','RH_3','T4','RH_4','T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8',
                        'T9','RH_9','T_out','Pressure','RH_out','Windspeed','Visibility','Tdewpoint','rv1','rv2'])
data_weights.index.name = 'Predictors'
data_weights.columns.name = 'Weights'
data_weights
```

```python
data_weights.plot(kind='bar')
plt.xlabel('Predictors')
plt.ylabel('Weights')
plt.show()
```
